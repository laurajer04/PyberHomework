```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
read_city = pd.read_csv(city_data_to_load)
read_ride = pd.read_csv(ride_data_to_load)
# Combine the data into a single dataset
merge_both = pd.merge(read_city, read_ride, on="city", how="left")
# Display the data table for preview
```

```python
merge_both.head()
```

```python
urban_avg = merge_both.loc[merge_both["type"] == "Urban"].groupby(['city'])['fare'].mean()
urban_rides = merge_both.loc[merge_both["type"] == "Urban"].groupby(['city']).count()['ride_id']
urban_drivers = merge_both.loc[merge_both["type"] == "Urban"].groupby(['city'])['driver_count'].value_counts()

sub_avg = merge_both.loc[merge_both["type"]=="Suburban"].groupby(['city'])['fare'].mean()
sub_rides = merge_both.loc[merge_both["type"]=="Suburban"].groupby(['city']).count()['ride_id']
sub_drivers = merge_both.loc[merge_both["type"]=="Suburban"].groupby(['city'])['driver_count'].value_counts()

rural_avg = merge_both.loc[merge_both["type"]=="Rural"].groupby(['city'])['fare'].mean()
rural_rides = merge_both.loc[merge_both["type"]=="Rural"].groupby(['city']).count()['ride_id']
rural_drivers = merge_both.loc[merge_both["type"]=="Rural"].groupby(['city'])['driver_count'].value_counts()

```

## Bubble Plot of Ride Sharing Data

```python
# Build the scatter plots for each city types
plt.scatter(urban_rides, urban_avg, s=urban_rides*10,
           marker = 'o', facecolors = "lightcoral", edgecolors = 'black', alpha = 0.5)
plt.scatter(sub_rides, sub_avg, s=sub_rides*10,
           marker = 'o', facecolors = "lightskyblue", edgecolors = 'black', alpha = 0.5)
plt.scatter(rural_rides, rural_avg, s=rural_rides*10,
           marker = 'o', facecolors = "gold", edgecolors= 'black', alpha = 0.5)

# Incorporate the other graph properties
plt.title('Pyber Ride Share Data 2016')
plt.xlabel('Total Number Rides (Per City)')
plt.ylabel('Average Fare ($)')
plt.grid(alpha=.7)

# Create a legend
plt.legend(['Urban', 'Suburban', 'Rural'])
# Incorporate a text label regarding circle size
plt.text(42,40, 'Note: Circle size correlates with driver count per city')
# Save Figure
plt.savefig('./PyberRideShare_Bubble.png')
```

## Total Fares by City Type

```python
# Calculate Type Percents
city_per = merge_both.groupby(['type']).sum()
city_data = city_per['fare']
# Build Pie Chart
plt.pie(city_data, explode=(0, 0,.1), labels=city_data.index, autopct='%1.1f%%', colors=['coral', 'lightskyblue','gold'], shadow=True, startangle=90,)
plt.title('% of Total Ride Fares by City Type')

# Save Figure
plt.savefig('./FaresByCity_Pie.png')
plt.show()
```

## Total Rides by City Type

```python
# Calculate Ride Percents
ride_per = merge_both.groupby(['type']).nunique()
ride_data = ride_per['fare']
# Build Pie Chart
plt.pie(ride_data, explode=(.1, .1,.1), labels=ride_data.index, autopct='%1.1f%%', colors=['orange', 'lightskyblue','cyan'], shadow=True, startangle=90,)
plt.title('% of Total Ride Fares by City Type')
# Save Figure
plt.savefig('./RideByCity_Pie.png')
plt.show()
```
